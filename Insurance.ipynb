Display Training Features : **Smoker**

```python
  df["smoker"] = df["smoker"].astype("category")
  df["smoker"] = df["smoker"].cat.codes
  df
```

Display Training Features : SEX

```python
  df["sex"] = df["sex"].astype("category")
  df["sex"] = df["sex"].cat.codes
  df
```

Display Training Features : **Region**

```python
  df["region"] = df["region"].astype("category")
  df["region"] = df["region"].cat.codes
  df
```

Check any null values....

```python
df.isnull().sum()
```

Multiple regression formula is Y = m1x1 + m2x2 + m3x3 + ........ mnxn

We are dropping the column "Charges" because it is represented by y in the above formula. i.e. the values of charges would depend on the values of x in the multiple regression.

Display X axis --- Independent Variables

```python
x = df.drop(columns = "charges")
x
```

Display Y axis i.e. the dependent variable

```python
y = df["charges"]
y
```

Now Using Train Test Split method to train the model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state=0)
```

Machine Learning model implementation

Using multiple LineaR Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

Training Dataset to the linear regression model

```python
lr.fit(x_train,y_train)
```

Displaying the Y intercept

```python
c = lr.intercept_
c
```

Displaying The coefficients

```python
m = lr.coef_
m
```

Build the training model and display the training data set prediction

```python
y_pred_train = lr.predict(x_train)
y_pred_train
```

Visualizations

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel("Original Charges")
plt.ylabel("Predicted Charges")
plt.show()
```

r2 score prediction accuracy. r2 score near 0 is the worst case scenario and r2 near 1 is the best case scenario.

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)
```

```python
y_pred_test = lr.predict(x_test)
```

Visualization of testing data set

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_test)
plt.xlabel("Original Charges")
plt.ylabel("Predicted Charges")
plt.show()
```

```python
r2_score(y_test,y_pred_test)
```

Difference between actual values and predicted values

```python
pred_y_df = pd.DataFrame({"Actual Value" : y_test, "Predicted Value" : y_pred_test, "Difference" : y_test - y_pred_test})
pred_y_df[0:25]
```
